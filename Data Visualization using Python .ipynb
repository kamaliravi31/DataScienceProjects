```python
#Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Basic Graph

x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Resize 
plt.figure(figsize=(5,3), dpi=300)
     

plt.plot(x,y, label='2x', color='red',linewidth=2, marker = '.',linestyle = '--')


#plt.plot(x,y,'b.--', label='2x') #simple short cut 

## Line Number Two
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'b', label='x2^2')
plt.plot(x2[5:], x2[5:]**2,'g^--', label='5 num')

plt.title('graph', fontdict = {'fontname' :'Comic Sans MS', 'fontsize': 20})
plt.xlabel('bbb')
plt.ylabel('kkk')

plt.xticks([0,2,3,4])
plt.yticks([0,2,4,6,8,7.5,7,10])

plt.legend()
#plt.savefig('mygraph', dpi=300)
plt.show()
```

```python
## BARCHARTS 

labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels,values)
patterns = ['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
#bars[0].set_hatch('/')
#bars[1].set_hatch('O')
#bars[2].set_hatch('*')
plt.figure(figsize=(6,4))

plt.show()

```

```python
# LINE GRAPH
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


gas = pd.read_csv('gas_prices.csv')
plt.figure(figsize=(12,5))

plt.plot(gas.Year, gas.Canada, 'b.-' , label = 'Canada')
plt.plot(gas.Year, gas.USA , 'r.-' ,label = 'USA')
plt.plot(gas.Year, gas['South Korea'] , 'g*--' ,label= 'SK')
plt.title('Gas prices over time in US Dollor')
plt.legend()
#iterate the year every 3 
plt.xticks(gas.Year[::3])


for country in gas:
    if country != 'Year':
        print(country)
        plt.plot(gas.Year, gas[country], marker='.')
        

plt.legend()
plt.show()
```

```python
## 
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
#Histograms 
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins = bins, color='#abcdef')
plt.xticks(bins)
plt.show()
```

```python
#Pie Charts

left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right= fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

labels = ['Left','Right']
colors = ['red','blue']
explode=(.1,.1)
plt.pie([left,right],labels=labels , colors = colors,  autopct='%.2f %%', pctdistance=0.5, explode=explode)
plt.show()
```

```python
# box and wisker plot

plt.style.use('ggplot')
plt.figure(figsize=(5,8))
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']


labels = ['b','m']
boxes=plt.boxplot([barcelona,madrid], labels=labels, patch_artist=True, medianprops = {'linewidth':2})

for box in boxes['boxes']:
    box.set(color='#abdecf', linewidth=2)
    box.set(facecolor='red') #patch_artist=true
plt.show()
```
